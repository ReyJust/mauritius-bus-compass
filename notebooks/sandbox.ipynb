```python
from typing import List
from unidecode import unidecode
import re
import pandas as pd

def normalize_str(df: pd.DataFrame, str_col: str)-> pd.DataFrame:

    df = df.copy()
    df[str_col] = df[str_col].str.lower()
    df[str_col] = df[str_col].apply(unidecode)
    df[str_col] = df[str_col].apply(lambda x: re.sub(r'[^\w\s]','', x))
    df[str_col] = df[str_col].str.strip()

    return df

normalized = normalize_str(df.dropna(subset='locality'), 'locality')
```
